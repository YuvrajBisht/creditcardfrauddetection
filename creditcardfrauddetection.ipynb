```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading dataset to pandas df
creditcard_data=pd.read_csv('/content/creditcard.csv.zip')
```

highly unbalanced dataset



0 - noramal transaction

```python
#separating the data for analysis
legit=creditcard_data[creditcard_data.Class == 0]
fraud=creditcard_data[creditcard_data.Class == 1]
```

Under-sampling

Build a sample dataset containing similar distribution of normal transaction and fraudalent transactions


no of fraudulent transaction = 492

taking 492 transaction from legit and joining it with 492 transaction of fraud.
This will make a good dataset

```python
legit_sample = legit.sample(n=492)
```

concat two df

```python
new_dataset = pd.concat([legit_sample , fraud], axis=0)
```

```python
new_dataset.head()
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

```python
x = new_dataset.drop(columns='Class',axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Splitting data into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model tarining

Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the logistic regression with training data
model.fit(x_train,y_train)
```

Find the performance our model

Model evaluation

Accuracy score

```python
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy)
```
